```python
import tensorflow as tf
import pandas as pd
import pathlib
import os
import matplotlib.pyplot as plt
import numpy as np
import csv
import scipy.interpolate as interp
import keras 
from keras import layers
import hickle as hkl
from keras.callbacks import History 
history = History()
import time as time
```

```python
dataset = hkl.load('inputimages.hkl') #load the input dataset 
model = keras.models.load_model('model_v2_images.h5')# load the model
np.shape(dataset)
```

```python
#this is a preliminary test with the calculation time
start = time.time()
outputval = model(np.array([dataset[0]]))
end = time.time()
elapsedtime = end-start
print ("elapsed calculation time: "+str(elapsedtime))
```

```python
# test result
plt.figure(figsize=(10,10))
plt.imshow (outputval[0])
```

```python
#this generates images for the first 5 circuits to check the model performance
i = [0,1,2,3,4]
for n in i:
    test = np.array(model((np.array([dataset[n]]))))
    plt.figure(figsize=(20,20))
    plt.imshow(test[0,:,:,0]+dataset[n]*0.05,cmap = 'Accent')
    plt.colorbar()
    plt.savefig('Results/circuit'+str(n)+'modelimage.png')
```
